### Convert `published_timestamp` to datetime format

To enable time-based operations and analysis, we convert the `published_timestamp` column to `datetime` format using `pd.to_datetime()`. This helps in tasks like filtering by year, extracting month/day, or plotting trends over time.


```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
```

# Find Out Number of Courses Per Subjects #

```python
plt.figure(figsize=(7,6))
sns.countplot(data = df , x = 'subject',hue = 'subject',edgecolor = 'black')
plt.tight_layout()
plt.xticks(rotation = 45)
plt.title('Numbers Of Courses Per Subject',fontsize = 13)
plt.xlabel('Subject',fontsize = 13)
plt.ylabel('Count',fontsize = 13)
plt.show()
```

# For Which Levels, Udemy Courses Providing The Courses #

```python
plt.figure(figsize=(7,6))
sns.countplot(data = df , x = 'level',hue = 'level',edgecolor = 'black')
plt.tight_layout()
plt.xticks(rotation = 45)
plt.title('Numbers Of Courses Per Level',fontsize = 13)
plt.xlabel('Level',fontsize = 13)
plt.ylabel('Count',fontsize = 13)
plt.show()
```

# Display The Count of Paid and Free Courses #

```python
sns.countplot(data = df , x = 'is_paid',hue = 'is_paid',edgecolor = 'black')
plt.tight_layout()
plt.title('Number Of Free And Paid Courses',fontsize = 13)
plt.xlabel('is_paid',fontsize = 13)
plt.ylabel('Count',fontsize = 13)
plt.show()
```

# Which Course Has More Lectures (Free or Paid)? #

```python
df.groupby('is_paid')['num_lectures'].mean()
```

# Which Courses Have A Higher Number of Subscribers Free or Paid? #

```python
df.groupby('is_paid')['num_subscribers'].mean().sort_values(ascending = False).plot(kind='bar',edgecolor = 'black')
plt.xlabel('is_paid',fontsize = 13)
plt.ylabel('num_subscribers',fontsize = 13)
plt.xticks(rotation = 0)
plt.tight_layout()
```

# Which Level Has The Highest Number of Subscribers? #

```python
df.groupby('level')['num_subscribers'].mean().sort_values(ascending = False).plot(kind='bar',edgecolor = 'black')
plt.xlabel('level',fontsize = 13)
plt.ylabel('num_subscribers',fontsize = 13)
plt.xticks(rotation = 0)
plt.tight_layout()
```
